```python
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily, Hourly
import pandas as pd


start = datetime(2019, 1, 1)
end = datetime(2021, 12, 31)

location = Point(55.9533, -3.1883, 70)

# Get hourly data for 2019 to 2021
data = Hourly(location, start, end)
data = data.fetch()

#Defining Weather Condition Code dictionary
coco_dict = {1:"Clear", 2:"Fair", 3:"Cloudy", 4:"Overcast", 5:"Fog",6:"Freezing Fog", 7:"Light Rain", 8:"Rain", 9:"Heavy Rain", 10:"Freezing Rain", 11:"Heavy Freezing Rain",12:"Sleet", 13:"Heavy Sleet",14:"Light Snowfall", 15:"Snowfall", 16:"Heavy Snowfall", 17:"Rain Shower", 18:"Heavy Rain Shower", 19:"Sleet Shower", 20:"Heavy Sleet Shower", 21:"Snow Shower",22:"Heavy Snow Shower", 23:"Lightening", 24:"Hail", 25:"Thunderstorm", 26:"Heavy Thunderstorm", 27:"Storm"}

df=pd.DataFrame(data)

#Deleting rows that hold garbage values
df.drop(["snow","tsun"], axis =1, inplace=True)

#Updating the coco column with values from the above delared dictionary
df["coco"].replace(coco_dict, inplace=True)



display(df)
df.to_csv('hourly_data_Edinburgh_code.csv')
```
